```python
import numpy as np 
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
x = np.array([[1], [2], [3], [4]]) 
y = np.array([2, 4, 6, 8])
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x, y)
y_pred_lin = lin_reg.predict(x)

lin_reg.coef_, lin_reg.intercept_
```

```python
Ridge = Ridge(alpha= 10.0)
Ridge.fit(x, y)
y_pred_Ridge = Ridge.predict(x)

Ridge.coef_, Ridge.intercept_
```

```python
lasso = Lasso(alpha= 10.0)
lasso.fit(x, y)
y_pred_lasso = lasso.predict(x)

lasso.coef_, lasso.intercept_
```

```python
print("Ridge Regression:")
print("MSE:", mean_squared_error(y, y_pred_lasso))
print("R2:", r2_score(y, y_pred_lasso))
print("==================================")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4)) 

plt.subplot(1, 3, 1) 

plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred_lin, color='red', label='Linear Regression')
plt.title('Linear Regression')
plt.legend()

plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso

X_reshaped = x.reshape(-1, 1) if x.ndim == 1 else x

model_lin = LinearRegression().fit(X_reshaped, y)
model_ridge = Ridge(alpha=1.0).fit(X_reshaped, y)
model_lasso = Lasso(alpha=1.0).fit(X_reshaped, y)

y_pred_lin = model_lin.predict(X_reshaped)
y_pred_ridge = model_ridge.predict(X_reshaped)
y_pred_lasso = model_lasso.predict(X_reshaped)

plt.figure(figsize=(10, 7))

plt.scatter(x, y, color='blue', alpha=0.6, label='Actual Data')

plt.plot(x, y_pred_lin, color='red', label='Linear Regression', linewidth=2)
plt.plot(x, y_pred_ridge, color='green', label='Ridge Regression', linestyle='--')
plt.plot(x, y_pred_lasso, color='purple', label='Lasso Regression', linestyle=':')

plt.title('Comparison of Linear, Ridge and Lasso Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
